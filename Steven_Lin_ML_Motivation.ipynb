```python
import pandas as pd
import numpy as np
```

```python
def string_to_array(string):
  """Converts a string to a numpy array where each letter is separated by comma."""
  array = np.array(list(string))
  return array
```

```python
url = 'https://github.com/cran2367/sgt/blob/25bf28097788fbbf9727abad91ec6e59873947cc/data/protein-uniprot-reviewed-Ano-10k.tab?raw=true'
df = pd.read_table(url, sep='\t')
```

```python
url = 'https://github.com/cran2367/sgt/blob/25bf28097788fbbf9727abad91ec6e59873947cc/data/protein_classification.csv?raw=true'
df1 = pd.read_csv(url)
```
